```python
import pandas as pd
import os
import re
from rapidfuzz import process, fuzz

# -------------------------------
# CONFIG: Paths
# -------------------------------
base_path = "/content/drive/MyDrive/UIDAII"
official_csv_path = os.path.join(base_path, "india_districts_master.csv")  # official list
output_path = os.path.join(base_path, "cleaned")  # folder to save cleaned CSVs
os.makedirs(output_path, exist_ok=True)

# List of all dataset files
files_to_process = [
    # Biometric
    "api_data_aadhar_biometric/api_data_aadhar_biometric_0_500000.csv",
    "api_data_aadhar_biometric/api_data_aadhar_biometric_500000_1000000.csv",
    "api_data_aadhar_biometric/api_data_aadhar_biometric_1000000_1500000.csv",
    "api_data_aadhar_biometric/api_data_aadhar_biometric_1500000_1861108.csv",

    # Demographic
    "api_data_aadhar_demographic/api_data_aadhar_demographic_0_500000.csv",
    "api_data_aadhar_demographic/api_data_aadhar_demographic_500000_1000000.csv",
    "api_data_aadhar_demographic/api_data_aadhar_demographic_1000000_1500000.csv",
    "api_data_aadhar_demographic/api_data_aadhar_demographic_1500000_2000000.csv",
    "api_data_aadhar_demographic/api_data_aadhar_demographic_2000000_2071700.csv",

    # Enrolment
    "api_data_aadhar_enrolment/api_data_aadhar_enrolment_0_500000.csv",
    "api_data_aadhar_enrolment/api_data_aadhar_enrolment_500000_1000000.csv",
    "api_data_aadhar_enrolment/api_data_aadhar_enrolment_1000000_1006029.csv"
]

# -------------------------------
# FUNCTION: Normalize names
# -------------------------------
def normalize_name(name):
    if pd.isna(name):
        return ""
    name = str(name).lower()
    name = re.sub(r'[^a-z0-9\s]', '', name)  # remove punctuation
    name = re.sub(r'\s+', ' ', name).strip()  # remove extra spaces
    return name

# -------------------------------
# FUNCTION: Fuzzy match district
# -------------------------------
def map_district(name, choices, threshold=90):
    if not name:
        return None
    match = process.extractOne(name, choices, scorer=fuzz.token_sort_ratio)
    if match and match[1] >= threshold:
        return match[0]
    return None

# -------------------------------
# STEP 1: Load official district list
# -------------------------------
official_df = pd.read_csv(official_csv_path)
official_df['district_norm'] = official_df['district'].apply(normalize_name)
official_df['state_norm'] = official_df['state_ut'].apply(normalize_name)

# Build lookup lists for fuzzy matching
official_districts = official_df['district_norm'].unique().tolist()

# -------------------------------
# STEP 2: Process each dataset
# -------------------------------
mapping_records = []
unmatched_records = []

for file_rel in files_to_process:
    file_path = os.path.join(base_path, file_rel)
    print(f"Processing: {file_path}")

    # Load dataset
    df = pd.read_csv(file_path)

    # Normalize dataset state/district
    df['district_norm'] = df['district'].apply(normalize_name)
    df['state_norm'] = df['state'].apply(normalize_name)

    # Map district to official district
    df['district_mapped_norm'] = df['district_norm'].apply(lambda x: map_district(x, official_districts))

    # Identify unmatched districts
    unmatched = df[df['district_mapped_norm'].isna()][['state', 'district']].drop_duplicates()
    if not unmatched.empty:
        unmatched_records.append(unmatched)
        print(f"‚ö†Ô∏è Unmatched districts in {os.path.basename(file_rel)}:")
        print(unmatched)

    # Merge official district names
    df = df.merge(
        official_df[['district_norm', 'district', 'state_ut']],
        left_on='district_mapped_norm',
        right_on='district_norm',
        how='left',
        suffixes=('', '_official')
    )

    # Final cleaned district column
    df['district_clean'] = df['district_official'].combine_first(df['district'])

    # Record mapping for audit
    mapping_records.append(
        df[['state', 'district', 'district_clean']].drop_duplicates()
    )

    # Save cleaned dataset
    cleaned_filename = os.path.join(output_path, os.path.basename(file_rel))
    df.to_csv(cleaned_filename, index=False)
    print(f"Saved cleaned file: {cleaned_filename}")

# -------------------------------
# STEP 3: Save district mapping CSV
# -------------------------------
mapping_df = pd.concat(mapping_records).drop_duplicates()
mapping_df.to_csv(os.path.join(output_path, "district_mapping.csv"), index=False)
print("Saved district mapping CSV.")

# -------------------------------
# STEP 3b: Save unmatched districts for manual review
# -------------------------------
if unmatched_records:
    unmatched_df = pd.concat(unmatched_records).drop_duplicates()
    unmatched_df.to_csv(os.path.join(output_path, "unmatched_districts.csv"), index=False)
    print(f"Saved unmatched districts CSV with {len(unmatched_df)} records.")
else:
    print("All districts matched successfully!")

# -------------------------------
# STEP 4: Verify district counts against official list
# -------------------------------

# Prepare official counts
official_counts = official_df.groupby('state_ut')['district'].nunique().reset_index()
official_counts.rename(columns={'district': 'official_district_count'}, inplace=True)

# Combine all cleaned files into one DataFrame for verification
all_cleaned_files = [os.path.join(output_path, os.path.basename(f)) for f in files_to_process]
combined_df = pd.concat([pd.read_csv(f) for f in all_cleaned_files], ignore_index=True)

# Compute cleaned district counts per state
cleaned_counts = combined_df.groupby('state_ut')['district_clean'].nunique().reset_index()
cleaned_counts.rename(columns={'district_clean': 'cleaned_district_count'}, inplace=True)

# Merge with official counts
verify_df = official_counts.merge(cleaned_counts, on='state_ut', how='left')

# Identify mismatches
verify_df['match'] = verify_df['official_district_count'] == verify_df['cleaned_district_count']
mismatches = verify_df[~verify_df['match']]

print("\n=== STATE / UT DISTRICT COUNT VERIFICATION ===")
print(verify_df)

if not mismatches.empty:
    print("\n‚ö†Ô∏è MISMATCHES FOUND:")
    for _, row in mismatches.iterrows():
        state = row['state_ut']
        official_count = row['official_district_count']
        cleaned_count = row['cleaned_district_count']
        print(f"- {state}: official={official_count}, cleaned={cleaned_count}")

    # Show missing districts per mismatched state
    for state in mismatches['state_ut']:
        official_districts_set = set(official_df[official_df['state_ut'] == state]['district'])
        cleaned_districts_set = set(combined_df[combined_df['state_ut'] == state]['district_clean'])
        missing_districts = official_districts_set - cleaned_districts_set
        if missing_districts:
            print(f"  Missing districts in {state}: {missing_districts}")
else:
    print("\n‚úÖ All states/UTs have correct district counts!")
```
